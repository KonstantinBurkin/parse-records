### Imports

```python
import pandas as pd
import json
import os, sys
from tqdm import tqdm

sys.path.append("../")

from scripts.parser_lab import LabParser
```

### table creation

```python
parser = LabParser()

records = []
json_dir = os.path.join("../", "data", "json_records", "DocsNew3")
for json_file in tqdm(os.listdir(json_dir)):
    path_json = os.path.join(json_dir, json_file)

    # read json
    with open(path_json, "r", encoding="utf-8") as file:
        data = json.load(file)

    table = parser.get_table_1(data=data)
    table_features = parser.table_to_features(table)
    gender = parser.get_sex(data)
    birthday = parser.get_age(data)

    data_in_record = {
        "record": json_file.removesuffix(".json"),
        "gender": gender,
        "birthday": birthday,
        **table_features,
    }

    records.append(data_in_record)

records_in_dataframe = pd.DataFrame(records)
```

```python
records_in_dataframe.head(3)
```

```python
for column in records_in_dataframe.columns:
    records_in_dataframe[column] = records_in_dataframe[column].astype(str)
```

```python
dir_path = "../data/parsed_dataframes"
os.makedirs(dir_path, exist_ok=True)
file_name = "lab_records.parquet"
file_path = os.path.join(dir_path, file_name)
records_in_dataframe.to_parquet(file_path, engine="pyarrow")
```
